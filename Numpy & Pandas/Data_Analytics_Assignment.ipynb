```python
import numpy as np 
import pandas as pd
from geopy.geocoders import Nominatim
import requests
import json
from timezonefinder import TimezoneFinder

```

**Task 1**

```python
geolocator = Nominatim(user_agent="find_city_coordinates")
cities = ["Budapest","New York","London","Mumbai","Cairo","Nairobi"]
```

```python
rows = []
column = []
time_zones = []
def sunrise_set(list_of_cities):
  for city in list_of_cities:
    
    this_row = []
    this_city = geolocator.geocode(city)
    obj = TimezoneFinder()
    time_zones.append(obj.timezone_at(lng=this_city.longitude, lat=this_city.latitude))
    query = f"https://api.sunrise-sunset.org/json?lat={this_city.latitude}&lng={this_city.longitude}&date=2022-06-01"
    resp = requests.get(url=query)
    resp = json.loads(resp.text)
    for key in resp['results'].keys():
      this_row.append(resp['results'][key])
      if key not in column:
        column.append(key)
    rows.append(this_row)

sunrise_set(cities)
print(time_zones)
```

```python
print(rows)
print(column)
```

```python
df = pd.DataFrame(rows, columns = column,index = cities)
df.info()
```

*   Create a new column which contains your datetime column localized to the US Eastern time zone.

```python
for zone in time_zones:
  df['solar_noon_US_Eastern'] = df['solar_noon'].dt.tz_localize(zone).dt.tz_convert('US/Eastern')
df
```
